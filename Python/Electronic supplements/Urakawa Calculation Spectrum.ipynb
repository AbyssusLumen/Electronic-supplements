```python
# =============================================================================
# COMPUTATION OF THE EIGENVALUES FOR t1=t2=2 AND k,l COPRIME BY URAKAWA
# =============================================================================

from sympy import symbols

k, l = symbols('k l', integer=True)
k=1                    #Urakawa computed the spectrum for arbitrary k,l but only for t1=t2=2
l=1

def d(z1,z2):          #Computation of the dimension of the representation (z1,z2)
    return (z1-z2+1)*(z1+2)*(z2+1)/2  
print("Results for k=", k, "l=", l,)
print("")
print("{:<10} {:<15} {:<10}".format("(z1, z2)", "η(z1,z2)", "mult(z1,z2)")) 
#print("{:<10} {:<15} {:<10}".format("(n_1, n_2)", "η(n_1,n_2)", "m(n_1,n_2)")) #to  to compare with the tables provided by Urakawa: uncomment this line and comment the line above
print("--------------------------------------------------------")

results = set()

for n_1 in range(0, 20):#n_1 and n_2 parameterize z1 and z2, count denotes how often the trivial representation occurs when restricted to S^1.
    for n_2 in range(0, 20):
        count = 0
        for P in range(n_1+1):
            for q in range(n_2+1):
                for r in range(P+n_2-q+1):
                    if k*n_1-l*n_2-(2*k+l)*P+(l-k)*q+(k+2*l)*r ==0  and 0 <= P <= n_1 and 0 <= q <= n_2 and 0 <= r <= P+n_2-q:
                        count += 1
        if count > 0:
            z1 = n_1 + n_2
            z2 = n_2
            result = (2/3) * (z1**2 + z2**2 - z1*z2 + 3*z1)
            results.add((z1,z2,f"({z1},{z2})", int(result), d(z1,z2)*count))
           # results.add((n_1, n_2, f"({n_1},{n_2})",int(result), count)) #to  to compare with the tables provided by Urakawa: uncomment this line and comment the line above
for r in sorted(results):
    print("{:<10} {:<15} {:<10}".format(r[2], r[3], r[4]))
```
